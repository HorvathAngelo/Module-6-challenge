```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import random
from citipy import citipy

# OpenWeatherMap API Key
weather_api_key = "5134fc12bdf38bff7941b48d060b1e8e"

# Set up a list to store your city data
city_data = []

# Generate random coordinates
lats = [random.uniform(-90, 90) for _ in range(500)]
lngs = [random.uniform(-180, 180) for _ in range(500)]

# Find nearest city for each coordinate pair
for lat, lng in zip(lats, lngs):
    city = citipy.nearest_city(lat, lng).city_name
    country = citipy.nearest_city(lat, lng).country_code

    # Check if the city is unique (avoid duplicates)
    if (city, country) not in [(data['City'], data['Country']) for data in city_data]:
        city_data.append({'City': city, 'Country': country, 'Lat': lat, 'Lng': lng})

# Convert the list of dictionaries to a DataFrame
city_df = pd.DataFrame(city_data)

# Function to get weather data for a city using the OpenWeatherMap API
def get_weather_data(city_name, country_code):
    base_url = "http://api.openweathermap.org/data/2.5/weather"
    params = {"q": f"{city_name},{country_code}", "appid": weather_api_key, "units": "metric"}

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an error for bad requests
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        print(f"Error occurred during API request for {city_name}: {e}")
        return None

# Loop through the cities and make API requests to fetch weather data
for index, row in city_df.iterrows():
    city_name = row['City']
    country_code = row['Country']

    weather_data = get_weather_data(city_name, country_code)

    if weather_data:
        city_df.at[index, 'Temperature (C)'] = weather_data['main']['temp']
        city_df.at[index, 'Humidity (%)'] = weather_data['main']['humidity']
        city_df.at[index, 'Cloudiness (%)'] = weather_data['clouds']['all']
        city_df.at[index, 'Wind Speed (m/s)'] = weather_data['wind']['speed']

# Create scatter plots to showcase the relationships between weather variables and latitude

# Scatter plot: Latitude vs. Temperature
plt.scatter(city_df['Lat'], city_df['Temperature (C)'], marker='o', edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Temperature (C)')
plt.title('City Latitude vs. Temperature')
plt.grid(True)
plt.show()

# Scatter plot: Latitude vs. Humidity
plt.scatter(city_df['Lat'], city_df['Humidity (%)'], marker='o', edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.title('City Latitude vs. Humidity')
plt.grid(True)
plt.show()

# Scatter plot: Latitude vs. Cloudiness
plt.scatter(city_df['Lat'], city_df['Cloudiness (%)'], marker='o', edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.title('City Latitude vs. Cloudiness')
plt.grid(True)
plt.show()

# Scatter plot: Latitude vs. Wind Speed
plt.scatter(city_df['Lat'], city_df['Wind Speed (m/s)'], marker='o', edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (m/s)')
plt.title('City Latitude vs. Wind Speed')
plt.grid(True)
plt.show()
```
