```python
import pandas as pd
import random

# Create a DataFrame with hypothetical hotel data
def generate_hotel_data(city_data):
    hotel_data = []
    for index, row in city_data.iterrows():
        lat = row["Lat"]
        lng = row["Lng"]
        hotel_name = f"Hotel {row['City']}"
        hotel_country = row["Country"]
        hotel_data.append({"City": row["City"], "Country": hotel_country, "Lat": lat, "Lng": lng, "Hotel Name": hotel_name})
    return pd.DataFrame(hotel_data)

# Replace this with your actual CSV file path
csv_file_path = "C:/Users/Horva/Downloads/Starter_Code (9)/Starter_Code/output_data/weather_data.csv"

# Load the CSV file into a DataFrame
city_data = pd.read_csv(csv_file_path)

# Generate hypothetical hotel data based on city coordinates
hotel_df = generate_hotel_data(city_data)

# Save the hotel data DataFrame as a CSV file in the same location as the cities CSV
hotel_data_csv = "hypothetical_hotel_data.csv"
hotel_df.to_csv(hotel_data_csv, index=False)

# Display the saved hotel data CSV file path
print(f"Hypothetical hotel data saved as {hotel_data_csv}")
```

```python
import pandas as pd
import folium
import requests

# Load the CSV files
city_data_df = pd.read_csv("C:/Users/Horva/Downloads/Starter_Code (9)/Starter_Code/output_data/cities.csv")
hotel_data_df = pd.read_csv("C:/Users/Horva/Downloads/Starter_Code (9)/Starter_Code/output_data/hotel_data_custom_name.csv")

# Create a map centered at the mean coordinates
m = folium.Map(location=[city_data_df['Lat'].mean(), city_data_df['Lng'].mean()], zoom_start=2)

# Add humidity points to the map
for index, row in city_data_df.iterrows():
    folium.CircleMarker(
        location=[row["Lat"], row["Lng"]],
        radius=row["Humidity"] / 10,  # Adjust the scaling as needed
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.5,
        popup=f"Humidity: {row['Humidity']}").add_to(m)

# Narrow down DataFrame based on ideal weather conditions
ideal_conditions_df = city_data_df[
    (city_data_df["Max Temp"] > 21)
    & (city_data_df["Max Temp"] < 27)
    & (city_data_df["Wind Speed"] < 4.5)
    & (city_data_df["Cloudiness"] == 0)
]

# Create a new DataFrame to store hotel information
hotel_df = ideal_conditions_df[["City", "Country", "Lat", "Lng", "Humidity"]].copy()

# Add hotel information from the hotel_data_df DataFrame
hotel_df["Hotel Name"] = hotel_data_df["Hotel Name"]

# Create a map for hotel information
hotel_map = folium.Map(location=[hotel_df['Lat'].mean(), hotel_df['Lng'].mean()], zoom_start=2)

# Add hotel markers to the map
for index, row in hotel_df.iterrows():
    folium.Marker(
        location=[row["Lat"], row["Lng"]],
        popup=f"City: {row['City']}\nCountry: {row['Country']}\nHotel Name: {row['Hotel Name']}",
        icon=folium.Icon(icon="cloud")
    ).add_to(hotel_map)

# Display the hotel map
hotel_map
```
