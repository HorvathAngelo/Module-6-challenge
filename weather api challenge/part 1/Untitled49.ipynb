```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

# Read the city data from the CSV file
city_data_df = pd.read_csv("C:/Users/Horva/Downloads/Starter_Code (9)/Starter_Code/output_data/cities.csv")

# Define a function to create scatter plots with linear regression
def create_linear_regression_plot(x_data, y_data, x_label, y_label, hemisphere, title):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x_data, y_data)
    regression_line = slope * x_data + intercept
    equation = f"y = {slope:.2f}x + {intercept:.2f}"
    r_squared = f"r^2 = {r_value**2:.2f}"
    
    # Create scatter plot with regression line
    plt.scatter(x_data, y_data, c='b', alpha=0.6, label='Data Points')
    plt.plot(x_data, regression_line, color='r', label='Linear Regression')
    
    # Annotate with regression equation and r-squared value
    plt.annotate(equation, (0.05, 0.9), xycoords='axes fraction', fontsize=12, color='r')
    plt.annotate(r_squared, (0.05, 0.85), xycoords='axes fraction', fontsize=12, color='r')
    
    # Set plot labels and title
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    
    # Add grid and legend
    plt.grid()
    plt.legend()
    
    # Show the plot
    plt.show()

# Separate data for Northern Hemisphere and Southern Hemisphere
northern_hemisphere = city_data_df[city_data_df["Lat"] >= 0]
southern_hemisphere = city_data_df[city_data_df["Lat"] < 0]

# Create scatter plots with linear regression for each relationship
create_linear_regression_plot(northern_hemisphere["Lat"], northern_hemisphere["Max Temp"], "Latitude", "Max Temp (C)", "Northern Hemisphere", "Northern Hemisphere: Temperature vs. Latitude")
create_linear_regression_plot(southern_hemisphere["Lat"], southern_hemisphere["Max Temp"], "Latitude", "Max Temp (C)", "Southern Hemisphere", "Southern Hemisphere: Temperature vs. Latitude")
create_linear_regression_plot(northern_hemisphere["Lat"], northern_hemisphere["Humidity"], "Latitude", "Humidity (%)", "Northern Hemisphere", "Northern Hemisphere: Humidity vs. Latitude")
create_linear_regression_plot(southern_hemisphere["Lat"], southern_hemisphere["Humidity"], "Latitude", "Humidity (%)", "Southern Hemisphere", "Southern Hemisphere: Humidity vs. Latitude")
create_linear_regression_plot(northern_hemisphere["Lat"], northern_hemisphere["Cloudiness"], "Latitude", "Cloudiness (%)", "Northern Hemisphere", "Northern Hemisphere: Cloudiness vs. Latitude")
create_linear_regression_plot(southern_hemisphere["Lat"], southern_hemisphere["Cloudiness"], "Latitude", "Cloudiness (%)", "Southern Hemisphere", "Southern Hemisphere: Cloudiness vs. Latitude")
create_linear_regression_plot(northern_hemisphere["Lat"], northern_hemisphere["Wind Speed"], "Latitude", "Wind Speed (m/s)", "Northern Hemisphere", "Northern Hemisphere: Wind Speed vs. Latitude")
create_linear_regression_plot(southern_hemisphere["Lat"], southern_hemisphere["Wind Speed"], "Latitude", "Wind Speed (m/s)", "Southern Hemisphere", "Southern Hemisphere: Wind Speed vs. Latitude")
```
